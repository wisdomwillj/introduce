```python
from selenium import webdriver
driver = webdriver.Chrome('D:\dataprac\data\chromedriver.exe')
url = 'http://www.kyobobook.co.kr/index.laf?orderClick=LJd'
driver.get(url)
```

```python
#//*[@id="searchKeyword"]
searchBook = input("책 이름 입력: ")
driver.find_element_by_xpath('//*[@id="searchKeyword"]').send_keys(searchBook)
import time
driver.find_element_by_xpath('//*[@id="searchTop"]/div/input').click()
time.sleep(2)
```

```python
def getList(soup):
    bCategory=[]
    bName=[]
    bAuthor=[]
    bPublisher=[]
    bPrice=[]
    bReview=[]
    searchList = soup.find('tbody', id='search_list')
    tr = searchList.find_all('tr')
    cnt = 0
    cnt_p = 0
    for i in tr:
        a = i.find('td', class_='detail')

        # 책분류
        category = a.find('span', class_='category2')
        bCategory.append(category.text.strip())

        # 책이름
        name = a.strong.text.strip()
        bName.append(name)
    
        # 책저자
        cnt+=1
        author = driver.find_element_by_xpath(f'//*[@id="search_list"]/tr[{cnt}]/td[2]/div[4]/a[1]') 
        bAuthor.append(author.text.strip())
    
        # 출판사
        cnt_p +=1
        publisher = driver.find_element_by_xpath(f'//*[@id="search_list"]/tr[{cnt_p}]/td[2]/div[4]/a[2]') 
        bPublisher.append(publisher.text.strip())
                    
        # 책가격
        b = i.find('td', class_='price')
        price = b.find('div', class_='org_price')
        bPrice.append(price.text.strip())
        
        # 책리뷰
        c = i.find('td', class_='info')
        label = c.b
        bReview.append(label.text.strip())
    return bCategory, bName, bAuthor, bPublisher, bPrice, bReview 
```

```python
from bs4 import BeautifulSoup
bookName = []
bookCategory = []
bookAuthor = []
bookPublisher = []
bookPrice = []
bookReview = []
# //*[@id="paging"]/li[1]/a
page_numbers = int(input("검색할 페이지 수를 입력: "))
for page in range(1,page_numbers):
    if page != 1:
        next_path = '//*[@id="paging"]/li[%s]/a'% page
        page_link = driver.find_element_by_xpath(next_path)
        page_link.click()
        time.sleep(3)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    bCategory, bName, bAuthor, bPublisher, bPrice, bReview = getList(soup)
    print(bName)
    bookCategory.extend(bCategory)
    bookName.extend(bName)
    bookPublisher.extend(bPublisher)
    bookReview.extend(bReview)
    bookAuthor.extend(bAuthor)
    bookPrice.extend(bPrice)
```

```python
import pandas as pd
import numpy as np
categroy_sr = pd.Series(bookCategory)
name_sr = pd.Series(bookName)
price_sr = pd.Series(bookPrice)
review_sr = pd.Series(bookReview)
author_sr =pd.Series(bookAuthor)
publisher_sr=pd.Series(bookPublisher)
to_df = pd.concat([categroy_sr,name_sr,author_sr, publisher_sr, price_sr,review_sr], axis=1)
to_df.columns = ['Category', 'Name', 'Author','Publisher', 'Price', 'Review']
to_df
```
